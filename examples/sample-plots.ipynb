```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import numpy as np
from math import sin, pi
import warnings

from kondo import use_style, make_subplots, color_ticks
```

```python
use_style()
```

```python
# LINE PLOT
def line_plot(ax):  
    rng = np.random.RandomState(4)
    x = np.linspace(0, 9, 500)
    y = np.cumsum(rng.randn(500, 10), 0)
    ax.set_title('Line Graph')
    ax.set_xlabel('— Time')
    ax.set_ylabel('— Random values')
    ax.set_xlim([0, 10])
    ax.set_ylim([-20, 60])
    ax.plot(x, y)
    ax.legend(['Bitcoin', 'Ethereum', 'Dollar', 'Oil'], title="Stonks")

# SCATTER PLOT
def scatter_plot(ax):
    rng = np.random.RandomState(4)
    x = np.linspace(0.6, pi-0.6, 100)
    y = [sin(x) + rng.rand() - 0.5 for x in x]
    t = np.linspace(-1, pi+0.2, 300)
    z = [0.5*sin(x*5) + rng.rand() - 0.5 for x in t]
    ax.set_title('Scatter Plot')
    ax.set_xlabel('— space')
    ax.set_ylabel('— altitude')
    plt.xlim([-0.2, pi+0.2])
    plt.ylim([-1.6, 1.6])
    ax.scatter(x, y, s=100, alpha=.6, color='C2')
    ax.scatter(t, z, s=100, alpha=.6, marker='^', color='C1')
    ax.legend(['sun', 'mountain'])

# DISTRIBUTIONS
def distribution_plot(ax):
    np.random.seed(2)
    data = np.random.multivariate_normal((0, 0), [(5, 2), (2, 2)], size=2000)
    data[:, 1] = np.add(data[:, 1], 2)
    ax.set_title('Distribution Plot')
    ax.set_xlabel('— Density')
    ax.set_ylabel('— Random values')
    ax.set_xlim([-10, 10])
    ax.set_ylim([0, 0.31])
    
    # supress seaborn FutureWarnings
    warnings.simplefilter(action='ignore', category=(FutureWarning, UserWarning))
    for col in range(2):
        sns.distplot(data[:, col], ax=ax, color='C' + str(col+3))

# POLAR PLOT
def polar_plot(ax):
    r = np.arange(0, 3.0, 0.01)
    theta = 2 * pi * r
    ax.plot(theta, r)
    ax.plot(0.5 * theta, r, ls='--')
    ax.set_title("Polar Axis Plot")
```

```python
def plot():
    fig, axes = make_subplots(2, 2, axes_w=8)
    line_plot(axes[0, 0])
    scatter_plot(axes[0, 1])
    distribution_plot(axes[1, 0])
    ax = plt.subplot(2, 2, 4, projection='polar')
    polar_plot(ax)
```

```python
plot()
```
